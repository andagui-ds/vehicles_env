# Análisis Exploratorio de Datos (EDA) - Mercado de Vehículos Usados 🚗

En este cuaderno realizaremos un análisis exploratorio preliminar sobre el conjunto de datos de anuncios de venta de coches en los EE.UU. El objetivo es comprender la estructura de los datos, identificar valores nulos y analizar las relaciones clave entre las variables antes de la construcción final de la aplicación web.

```python
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display

# Cargar el conjunto de datos
# Como el cuaderno EDA.ipynb está dentro de la carpeta 'notebooks', usamos '../' para buscar el CSV en la raíz
df = pd.read_csv('../vehicles_us.csv')

# Mostrar información básica del DataFrame
print(f"Dimensiones del dataset: {df.shape[0]} filas y {df.shape[1]} columnas.\n")
df.info()
```

```python
# Crear un botón interactivo usando ipywidgets
boton_hist = widgets.Button(
    description='Construir Histograma',
    button_style='info', # 'success', 'info', 'warning', 'danger' o ''
    tooltip='Haz clic para generar el histograma de precios',
    icon='bar-chart'
)

salida_hist = widgets.Output()

def al_hacer_clic_hist(b):
    with salida_hist:
        salida_hist.clear_output() # Limpia el gráfico anterior si existía
        print("Generando distribución de precios...")
        # Crear histograma con Plotly Express
        fig_hist = px.histogram(
            df, 
            x='price', 
            title='Distribución General de Precios de Venta',
            labels={'price': 'Precio ($)'},
            color_discrete_sequence=['#1f77b4']
        )
        fig_hist.update_layout(template="plotly_white")
        fig_hist.show()

# Conectar el botón con la función lógica
boton_hist.on_click(al_hacer_clic_hist)

# Mostrar los componentes en el cuaderno
display(boton_hist, salida_hist)
```

```python
# Crear una casilla de verificación interactiva
casilla_disp = widgets.Checkbox(
    value=False,
    description='Mostrar gráfico de dispersión',
    disabled=False,
    indent=False
)

salida_disp = widgets.Output()

def al_cambiar_casilla(change):
    with salida_disp:
        salida_disp.clear_output()
        if change['new']: # Si la casilla está marcada (True)
            print("Generando relación entre Kilometraje y Precio...")
            # Crear scatter plot con Plotly Express
            fig_scatter = px.scatter(
                df, 
                x='odometer', 
                y='price', 
                color='condition',
                title='Relación: Kilometraje (Odómetro) vs Precio',
                labels={'odometer': 'Odómetro (Millas)', 'price': 'Precio ($)', 'condition': 'Condición'},
                opacity=0.5
            )
            fig_scatter.update_layout(template="plotly_white")
            fig_scatter.show()
        else:
            print("Gráfico ocultado. Marca la casilla para mostrarlo.")

# Conectar la casilla con la función lógica
casilla_disp.observe(al_cambiar_casilla, names='value')

# Mostrar los componentes en el cuaderno
display(casilla_disp, salida_disp)
```
